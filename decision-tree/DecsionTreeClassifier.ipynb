```python
# import libraries
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree
import pandas as pd
```

```python
df = pd.read_csv('C:\cygwin64\home\salis\data\caesarian.csv')
```

```python
df.head()
```

```python
len(df)
```

```python
# column has a space in front of it
df.rename(columns={" caesarian": "caesarian"}, inplace=True)
```

```python
rs = 23
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('caesarian', axis=1), 
                                                    df['caesarian'], test_size=0.6, random_state=rs)
```

```python
X_train.head()
```

```python
dtclf2 = DecisionTreeClassifier(random_state=rs, max_depth=2)
dtclf4 = DecisionTreeClassifier(random_state=rs, max_depth=4)
```

```python
dtclf2.fit(X_train,y_train)
```

```python
dtclf4.fit(X_train,y_train)
```

```python
y2_predict = dtclf2.predict(X_test)
```

```python
y4_predict = dtclf4.predict(X_test)
```

```python
accuracy_score(y_test, y2_predict)
```

```python
accuracy_score(y_test, y4_predict)
```

```python
fig= plt.figure(figsize=(8,8))
tree.plot_tree(dtclf2)  
```

```python
fig= plt.figure(figsize=(14,14))
tree.plot_tree(dtclf4)  
```

## Why Vary Depth?



You vary the depth of a decision tree to tune the model. Not going deep enough may underfit the data, while going too deep will overfit the data.
